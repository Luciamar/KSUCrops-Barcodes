- **Importing the modules**

```python
import os
import glob
import pandas as pd
from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
from reportlab.graphics.barcode import code128
```

<br>

# **Plot stakes**

- **Code for the** `...stake_labels.csv`

```python
path = str(os.getcwd())+"/1. Labels to be done/"


for filepath in glob.glob(os.path.join(path, '*stake_labels.csv')):

    
    database = pd.read_csv(filepath)
    
    filepath = str(filepath)
    page_width = 4 ; page_height = 1
    page = canvas.Canvas("2. Labels already done/"+filepath[len(path):-4]+".pdf")
    page.setPageSize(size=(page_width*inch,
                           page_height*inch))
    
    
    for i in range(0,len(database)):
        
        
        page.rect(0.05*inch, (0.05-0.035)*inch, 3.95*inch, 0.9*inch, stroke=1, fill=0)
        page.rect(2.875*inch, (0.05-0.035)*inch, 1.175*inch, 0.9*inch, stroke=1, fill=0)
        page.rect(1.65*inch, (0.05-0.035)*inch, 1.175*inch, 0.9*inch, stroke=1, fill=0)
        page.rect(0.425*inch, (0.05-0.035)*inch, 1.175*inch, 0.9*inch, stroke=1, fill=0)
        page.rect(2.875*inch, (0.05-0.035+0.475)*inch, 1.175*inch, (0.9-0.475)*inch, stroke=1, fill=1)
        page.rect(0*inch, (0.05-0.035)*inch, 0.425*inch, 0.9*inch, stroke=1, fill=1)
        
        
        page.setFont('Helvetica-Bold', 25)
        page.setFillColorRGB(255,255,255)
        page.drawCentredString((page_width - (1.125/2))*inch, (0.58)*inch,
                               str(database.loc[i,'stake_info6']))
    
        page.setFont('Helvetica-Bold', 10)
        page.setFillColorRGB(0,0,0)
        page.drawCentredString(2.2375*inch, (0.71)*inch,
                               str(database.loc[i,'stake_info1']))
        
        page.setFont('Helvetica', 8)
        page.drawCentredString(2.2375*inch, (0.56)*inch,
                               str(database.loc[i,'stake_info2']))
        
        page.setFont('Helvetica', 8)
        page.drawCentredString(2.2375*inch, (0.41)*inch,
                               str(database.loc[i,'stake_info3']))
        
        page.setFont('Helvetica', 8)
        page.drawCentredString(2.2375*inch, (0.26)*inch,
                               str(database.loc[i,'stake_info4']))
        
        page.setFont('Helvetica-Bold', 7)
        page.drawCentredString(2.2375*inch, (0.11)*inch,
                               str(database.loc[i,'stake_info5']))
    
    
        b_code128 = code128.Code128(str(database.loc[i,'stake_info6']),
                                    barHeight = 0.3*inch, barWidth = 0.8)
        
        b_code128.lquiet = 0; b_code128.rquiet = 0 
        b_code_start = (page_width - (1.125/2)) - (b_code128.width/inch)/2
        b_code128.drawOn(page, b_code_start*inch, (0.13-0.035)*inch)
        
        
        page.showPage()
        
    page.save()
```

# **Soil sampling**

- **Code for the** `...soil_labels.csv`

```python
path = str(os.getcwd())+"/1. Labels to be done/"


for filepath in glob.glob(os.path.join(path, '*soil_labels.csv')):

    
    database = pd.read_csv(filepath)
    
    filepath = str(filepath)
    page_width = 2.25 ; page_height = 1.25
    page = canvas.Canvas("2. Labels already done/"+filepath[len(path):-4]+".pdf")
    page.setPageSize(size=(page_width*inch,
                           page_height*inch))
    
    
    for i in range(0,len(database)):
        
        
        page.rect(0.05*inch, (0.05-0.035)*inch, 2.15*inch, 1.15*inch, stroke=1, fill=0)
    
    
        page.setFont('Helvetica', 6)
        page.drawCentredString((1*(page_width/4))*inch, 1.02*inch,
                               str(database.loc[i,'soil_info1']))
        
        page.setFont('Helvetica-Bold', 6)
        page.drawCentredString((2*(page_width/4))*inch, 1.02*inch,
                               str(database.loc[i,'soil_info2']))
        
        page.setFont('Helvetica', 6)
        page.drawCentredString((3*(page_width/4))*inch, 1.02*inch,
                               str(database.loc[i,'soil_info3']))
        
        page.setFont('Helvetica-Bold', 10)
        page.drawCentredString((1*(page_width/3)-0.1)*inch, 0.77*inch,
                               str(database.loc[i,'soil_info4']))
        
        page.setFont('Helvetica', 10)
        page.drawCentredString((2*(page_width/3)+0.05)*inch, 0.77*inch,
                               str(database.loc[i,'soil_info5']))
        
        page.setFont('Helvetica-Bold', 14)
        page.drawCentredString((page_width/2)*inch, 0.48*inch,
                               str(database.loc[i,'soil_info6']))
        
    
        b_code128 = code128.Code128(database.loc[i,'soil_info6'],
                                    barHeight = 0.25*inch, barWidth = 0.85)
        
        b_code128.lquiet = 0; b_code128.rquiet = 0
        b_code_start = (page_width/2) - (b_code128.width/inch)/2    
        b_code128.drawOn(page, b_code_start*inch, 0.09*inch)
        
        page.showPage()
        
    page.save()
```

# **Biomass sampling**

- **Code for the** `...biomass_labels.csv`

```python
path = str(os.getcwd())+"/1. Labels to be done/"


for filepath in glob.glob(os.path.join(path, '*biomass_labels.csv')):

    
    database = pd.read_csv(filepath)
    
    filepath = str(filepath)
    page_width = 3 ; page_height = 2
    page = canvas.Canvas("2. Labels already done/"+filepath[len(path):-4]+".pdf")
    page.setPageSize(size=(page_width*inch,
                           page_height*inch))
    
    
    for i in range(0,len(database)):
        
        
        page.rect(0.05*inch, (0.05-0.025)*inch, 2.9*inch, 1.9*inch, stroke=1, fill=0)
    
    
        page.setFont('Helvetica-Bold', 14)
        page.drawCentredString(1.5*inch, 1.6*inch,
                               str(database.loc[i,'biomass_info1']))
    
        page.setFont('Helvetica-Bold', 12)
        page.drawCentredString(1.5*inch, 1.2*inch,
                               str(database.loc[i,'biomass_info2']))
    
        page.setFont('Helvetica', 10)
        page.drawCentredString(1.5*inch, 0.9*inch,
                               str(database.loc[i,'biomass_info3']))
    
    
        b_code128 = code128.Code128(database.loc[i,'biomass_info1'],
                                    barHeight = 0.4*inch, barWidth = 0.7)
        
        b_code128.lquiet = 0; b_code128.rquiet = 0
        b_code_start = (page_width/2) - (b_code128.width/inch)/2   
        b_code128.drawOn(page, b_code_start*inch, 0.3*inch)
        
        
        if pd.isnull(database.loc[i,'biomass_ucode']) == False:
            
            page.setFont('Helvetica-Bold', 8)
            page.drawCentredString(1.5*inch, 0.15*inch,
                                   str(database.loc[i,'biomass_ucode']))
        
        page.showPage()
        
    page.save()
```

# **Plant dry tissue**

- **Code for the** `...tissue_labels.csv`

```python
path = str(os.getcwd())+"/1. Labels to be done/"


for filepath in glob.glob(os.path.join(path, '*tissue_labels.csv')):

    
    database = pd.read_csv(filepath)
    
    filepath = str(filepath)
    page_width = 1.25 ; page_height = 2.25
    page = canvas.Canvas("2. Labels already done/"+filepath[len(path):-4]+".pdf")
    page.setPageSize(size=(page_width*inch,
                           page_height*inch))
    
    
    for i in range(0,len(database)):
        
        
        page.rect((0.05-0.035)*inch, 0.75*inch, 1.15*inch, 1.45*inch, stroke=1, fill=0)
        page.rect((0.05-0.035)*inch, 1.04*inch, 1.15*inch, 0.65*inch, stroke=1, fill=1)
    
    
        page.setFont('Helvetica-Bold', 9)
        page.drawCentredString(((page_width/2)-0.035)*inch, 2*inch,
                               str(database.loc[i,'tissue_info1']))
    
        page.setFont('Helvetica-Bold', 10)
        page.drawCentredString(((page_width/2)-0.035)*inch, 1.8*inch,
                               str(database.loc[i,'tissue_info2']))
    
        page.setFont('Helvetica-Bold', 20)
        page.setFillColorRGB(255,255,255)
        page.drawCentredString(((page_width/2)-0.035)*inch, 1.4*inch,
                               str(database.loc[i,'tissue_info3']))
    
        page.setFont('Helvetica-Bold', 14)
        page.setFillColorRGB(255,255,255)
        page.drawCentredString(((page_width/2)-0.18-0.035)*inch, 1.15*inch,
                               str(database.loc[i,'tissue_info4']))
    
        page.setFont('Helvetica', 14)
        page.setFillColorRGB(255,255,255)
        page.drawCentredString(((page_width/2)+0.18-0.035)*inch, 1.15*inch,
                               str(database.loc[i,'tissue_info5']))
        
        page.setFont('Helvetica-Bold', 9)
        page.setFillColorRGB(0,0,0)
        page.drawCentredString(((page_width/2)-0.035)*inch, 0.85*inch,
                               str(database.loc[i,'tissue_info6']))
    
    
        b_code128 = code128.Code128(database.loc[i,'tissue_ucode'],
                                    barHeight = 0.3*inch, barWidth = 0.85)
        
        b_code128.lquiet = 0; b_code128.rquiet = 0
        b_code_start = (page_width/2) - (b_code128.width/inch)/2    
        b_code128.drawOn(page, (b_code_start-0.035)*inch, 0.35*inch)
        
        page.setFont('Helvetica', 7)
        page.setFillColorRGB(0,0,0)
        page.drawCentredString(((page_width/2)-0.035)*inch, 0.2*inch,
                               str(database.loc[i,'tissue_ucode']))
        
        page.setFont('Helvetica-Bold', 8)
        page.setFillColorRGB(0,0,0)
        page.drawCentredString(((page_width/2)-0.035)*inch, 0.075*inch,
                               str("Original"))
        
        page.showPage()
        
        
        page.rect((0.05-0.035)*inch, 0.75*inch, 1.15*inch, 1.45*inch, stroke=1, fill=0)
        page.rect((0.05-0.035)*inch, 1.04*inch, 1.15*inch, 0.65*inch, stroke=1, fill=0)
    
    
        page.setFont('Helvetica-Bold', 9)
        page.drawCentredString(((page_width/2)-0.035)*inch, 2*inch,
                               str(database.loc[i,'tissue_info1']))
    
        page.setFont('Helvetica-Bold', 10)
        page.drawCentredString(((page_width/2)-0.035)*inch, 1.8*inch,
                               str(database.loc[i,'tissue_info2']))
    
        page.setFont('Helvetica-Bold', 20)
        page.drawCentredString(((page_width/2)-0.035)*inch, 1.4*inch,
                               str(database.loc[i,'tissue_info3']))
    
        page.setFont('Helvetica-Bold', 14)
        page.drawCentredString(((page_width/2)-0.18-0.035)*inch, 1.15*inch,
                               str(database.loc[i,'tissue_info4']))
    
        page.setFont('Helvetica', 14)
        page.drawCentredString(((page_width/2)+0.18-0.035)*inch, 1.15*inch,
                               str(database.loc[i,'tissue_info5']))
        
        page.setFont('Helvetica-Bold', 9)
        page.drawCentredString(((page_width/2)-0.035)*inch, 0.85*inch,
                               str(database.loc[i,'tissue_info6']))
    
    
        b_code128 = code128.Code128(database.loc[i,'tissue_ucode'],
                                    barHeight = 0.3*inch, barWidth = 0.85)
        
        b_code128.lquiet = 0; b_code128.rquiet = 0
        b_code_start = (page_width/2) - (b_code128.width/inch)/2    
        b_code128.drawOn(page, (b_code_start-0.035)*inch, 0.35*inch)
        
        page.setFont('Helvetica', 7)
        page.setFillColorRGB(0,0,0)
        page.drawCentredString(((page_width/2)-0.035)*inch, 0.2*inch,
                               str(database.loc[i,'tissue_ucode']))
        
        page.setFont('Helvetica-Bold', 8)
        page.setFillColorRGB(0,0,0)
        page.drawCentredString(((page_width/2)-0.035)*inch, 0.075*inch,
                               str("Duplicated"))
        
        page.showPage()
        
    page.save()
```
